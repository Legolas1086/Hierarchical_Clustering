```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

Importing data

```python
data=pd.read_csv('Mall_Customers.csv')
print(data.head())
```

```python
print(data.info())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,3:5].values
```

Determinig optimum number of clusters through Dendrogram

```python
from scipy.cluster import hierarchy
dendro=hierarchy.dendrogram(hierarchy.linkage(x,method='ward'))
plt.xlabel('age')
plt.ylabel('salary')
plt.title('Dendrogram')
plt.show()
```

Model Build

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
y=ac.fit_predict(x)
print(y)
```

Plotting the Clusters

```python
plt.scatter(x[y==0,0],x[y==0,1],color='red',label='Category 1')
plt.scatter(x[y==1,0],x[y==1,1],color='orange',label='Category 2')
plt.scatter(x[y==2,0],x[y==2,1],color='green',label='Category 3')
plt.scatter(x[y==3,0],x[y==3,1],color='blue',label='Category 4')
plt.scatter(x[y==4,0],x[y==4,1],color='pink',label='Category 5')
plt.xlabel('age')
plt.ylabel('salary')
plt.legend()
plt.show()
```
